```python
total_count = 0
data2 = np.load(f'data/combs_iters2000.npz', allow_pickle=True)

counter = 0
for j in range(10):
    combinations, initializations, misclassified, classes, weights, biases, losses, activations, perfect_iters = \
        [], [], [], [], [], [], [], [], []
    for i in range(10):
        data = np.load(f'data/xor_iters2000_job{j*10+i}.npz', allow_pickle=True)
#         combinations += data2['comb'].tolist()
        initializations += data['initial'].tolist()
        misclassified += data['miscl'].tolist()
        classes += data['cl'].tolist()
        weights += data['w'].tolist()
        biases += data['b'].tolist()
        losses += data['l'].tolist()
        activations += data['act'].tolist()
        perfect_iters += data['pi'].tolist()
    combinations = data2['combs'].tolist()[counter:counter+len(initializations)]
    counter += len(initializations)
    df = pd.DataFrame({'combs': combinations, 'inits': initializations, 'misclass': misclassified, 'classes': classes,
                  'weights': weights, 'biases': biases, 'losses': losses, 'acts': activations, 'perfect_iters': perfect_iters})
    df['combs_tuple'] = [tuple(x) for x in df['combs']]
#     grouped = df.groupby('combs_tuple')
#     for key, item in grouped:
#         print(grouped.get_group(key), "\n\n")
    good_ones = df.groupby('combs_tuple').filter(lambda x: x['perfect_iters'].notna().all())
    good_ones['mean_iters'] = good_ones.groupby('combs_tuple').mean()['perfect_iters']
#     print(good_ones.groupby('combs_tuple').mean()['perfect_iters'])
    print(good_ones.groupby('combs_tuple').agg(['mean', 'median', 'min', 'max', len])['perfect_iters'])
#     print(good_ones.groupby('combs_tuple').min()['perfect_iters'])
#     print(good_ones.groupby('combs_tuple').first()['combs'] if len(good_ones['combs']) > 0 else "")
    print(100*'x')

    for kombinace in [(1, 2, -2, -1, 1, 2), (1, 2, 3, 3, -1, 3), (1, 1, 2, 3, 1, 4), (1, 2, 3, 4, 1, 5)]:
        if kombinace in good_ones.groupby('combs_tuple').groups.keys():
            found_group = good_ones.groupby('combs_tuple').get_group(kombinace)
            print(found_group['weights'].iloc[0])
            print(found_group['biases'].iloc[0])

    total_count+=len(good_ones.groupby('combs_tuple').first()['combs'].index)
print(total_count)
```
